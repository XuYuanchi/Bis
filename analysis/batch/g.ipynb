```python
import scanpy as sc
import numpy as np
import pandas as pd
import h5py
import scib
import scipy.io as sio
```

```python
sc.settings.verbosity = 1             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_versions()
sc.settings.set_figure_params(dpi=200, frameon=False, figsize=(3,3), facecolor='white')
```

```python
# load raw data
adata = sc.read_h5ad('/home/suyanchi/project/dab/data/batch/human_pancreas.h5ad')
adata
```

```python
methods = ['alra', 'dca', 'deepimpute', 'magic', 'saver', 'scimpute', 'scscope', 'scvi']
dataset_name = 'human_pancreas'
```

```python
i=0 # data
j=7# methods
re=0.1
print(methods[j])
```

```python
if j in [4,5]:
    dat_sc = sio.loadmat('/home/suyanchi/project/dab/results/batch/'+dataset_name+'_'+methods[j]+'.mat')['data']
else:
    f = h5py.File('/home/suyanchi/project/dab/results/batch/'+dataset_name+'_'+methods[j]+'.h5')
    dat_sc = f['data']
    f.close
    dat_sc = np.array(dat_sc)
if dat_sc.shape!=adata.shape:
    dat_sc = dat_sc.T
adata.X = dat_sc
adata
```

```python
scib.pp.reduce_data(adata, batch_key='tech')
# sc.tl.pca(adata)
# sc.external.pp.bbknn(adata, batch_key='tech')
sc.external.pp.harmony_integrate(adata, key='tech')
# sc.external.pp.mnn_correct(adata, batch_key='tech')
```

```python
sc.pp.neighbors(adata, use_rep='X_pca_harmony')
# sc.pp.neighbors(adata)
```

```python
sc.tl.umap(adata)
```

```python
sc.pl.umap(adata,color=['tech', 'celltype'])
```

```python
sc.tl.leiden(adata, resolution=0.2)
sc.pl.umap(adata, color=['leiden', 'tech', 'celltype'])
```

```python
ari = scib.metrics.ari(adata, group1 = 'leiden', group2='celltype')
nmi = scib.metrics.nmi(adata, group1 = 'leiden', group2='celltype')
asw = scib.metrics.silhouette(adata, label_key='celltype', embed='X_pca_harmony')
b_asw = scib.metrics.silhouette_batch(adata, batch_key='tech', label_key='celltype', embed='X_pca_harmony', verbose=False)
print('ARI: {:.4f}, NMI: {:.4f}, ASW: {:.4f}, batch-ASW: {:.4f}'.format(ari, nmi, asw, b_asw))
```

```python
df = pd.read_csv('/home/suyanchi/project/dab/results/batch/'+dataset_name+'_metric.csv', header=0, index_col=0)
df
```

```python
# combineing results 
df1 = pd.DataFrame({'Methods': [methods[j]],
                   'ARI': [ari],
                   "NMI": [nmi],
                   "ASW": [asw],
                   "b_ASW": [b_asw]})
df = pd.concat([df, df1])
df
```

```python
df.to_csv('/home/suyanchi/project/dab/results/batch/'+dataset_name+'_metric.csv', sep=',')
```
